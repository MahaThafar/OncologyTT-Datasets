## Remove duplicates rows and write dataframes [DB_ID, DB_DR_NAME, TargetID, GeneName]

```python
for cancer in comCancers:
    print(cancer)
    fileName = 'Data/TTD/approved_DT_not_clean_with_DUPLICATES/DT_'+str(cancer)+'.xlsx'
    TTD_cancer = pd.read_excel(fileName)
    print(TTD_cancer.shape)
    TTD_cancer_no_dup = TTD_cancer.drop_duplicates(subset=['DB_ID', 'TTD TargetID','Gene Name'])
    print(TTD_cancer_no_dup.shape)
    TTD_cancer_no_dup.to_excel('Data/TTD/Approved_DTIs_info_TTD/DT_'+str(cancer)+'_all_info.xlsx',index = False)
    
    subset = TTD_cancer_no_dup[['DB_ID','DB_DR Name','TTD TargetID','Gene Name']].copy()
    subset.to_excel('Data/TTD/approvedDT/DT_'+str(cancer)+'.xlsx',index = False)
```

```python
for cancer in comCancers:
    
    file_syn = 'Data/Cancer_dr_synonums/'+str(cancer)+'_DB_syn.csv'
    dr_DB = pd.read_csv(file_syn)
    
    print('\n## Data curation for '+str(cancer)+' cancer: ##')
    Dr_file =  "Data/Approved_cancer_drugs_list/"+str(cancer)+"_approved_drugs.txt"
    app_dr_list = np.genfromtxt(Dr_file, delimiter='\n',dtype=str)
    
    approved_DT_dic = {}
    approved_DT = []
    c1=0
    c_all=0
    
    for nci, dr, DB_id in zip(app_dr_list, dr_DB['DrugBank Drug Name'],dr_DB['DrugBank ID']):
        for i in range(1,TTD_DT.shape[0]):
            
            if str(dr) in str(TTD_DT['Drug_Name'][i]) or str(TTD_DT['Drug_Name'][i]) in str(dr):
                
                drug_syn = dr_DB['All Synonyms'].str.split('[')
                #print(drug_syn)
                if TTD_DT['Drug_Name'][i] in (dr or str(drug_syn)):
                    it0 = TTD_DT.loc[i][0]
                    it1 = TTD_DT.loc[i][1]
                    it2 = TTD_DT.loc[i][2]
                    it3 = TTD_DT.loc[i][3]
                    it4 = TTD_DT.loc[i][4]
                    it5 = TTD_DT.loc[i][5]
                    drug_info = DB_id, dr, nci, it0, it1, it2, it3, it4, it5
                    
                    if(TTD_DT['Highest_status'][i]=='Approved'):
                        approved_DT.append(drug_info)
                        #approved_DT_dic(dr)=TTD_DT[:][i]
                        c1 = c1+1

                    c_all = c_all+1
    
    approved_DT_df = pd.DataFrame(approved_DT, columns=['DB_ID','DB_DR Name','NCI_DR Name',
                                                        'TTD TargetID','Target_Name',
                                                        'TTD DrugID','Drug_Name',
                                                        'Highest_status','MOA'])
    
    #Save files of approved DT & other DT stat
    approved_DT_df.to_excel('Data/TTD/approvedDT/DT_'+str(cancer)+'.xlsx',index = False)
    
    print('number of all approved drugs for cancer from NCI:', len(app_dr_list))
    print('\nNumber of *all drugs found in TTD* data for '+str(cancer)+' cancer =', c_all)
    print('Number of *approved* DT with for '+str(cancer)+' cancer =', c1)
    print('\n_________________________________________________________________________________')
```
